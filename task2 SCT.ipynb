```python
import pandas as pd
# Load the dataset
data = pd.read_csv("C:\\Users\\amish\\Downloads\\Titanic-Dataset.csv")
# Display the first few rows
print(data.head())
```

```python
# Calculate the survival rate
survival_rate = data['Survived'].mean()
print(f'Survival Rate: {survival_rate * 100:.2f}%')
```

```python
# Calculate the average age
average_age = data['Age'].mean()
print(f'Average Age: {average_age:.2f}')
```

```python
# Count male and female passengers
gender_counts = data['Sex'].value_counts()
print(gender_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Visualize survival rates by gender
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title('Survival Counts by Gender')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=data,color="green")
plt.title('Survival Rate by Gender')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Survived', data=data, palette='viridis')
plt.title('Survival Rate by Passenger Class', pad=20, fontweight='bold')
plt.ylabel('Survival Probability')
plt.xlabel('Passenger Class (1 = First, 2 = Second, 3 = Third)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='Age', data=data, palette=['#ff6961', '#77dd77'], split=True)
plt.title('Age Distribution by Survival Status', pad=20, fontweight='bold')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.ylabel('Age')
plt.xlabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=data, palette='pastel')
plt.title('Fare Distribution by Passenger Class', pad=20, fontweight='bold')
plt.ylabel('Fare (in currency units)')
plt.xlabel('Passenger Class')
plt.ylim(0, 300)  
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Embarked', hue='Survived', data=data, palette=['#ff6961', '#77dd77'])
plt.title('Survival Count by Embarkation Port', pad=20, fontweight='bold')
plt.legend(['Did Not Survive', 'Survived'])
plt.xlabel('Embarkation Port (C = Cherbourg, Q = Queenstown, S = Southampton)')
plt.ylabel('Passenger Count')
plt.show()
```

```python
family_data = data[['SibSp', 'Parch', 'Survived']].copy()
family_data['FamilySize'] = family_data['SibSp'] + family_data['Parch']
plt.figure(figsize=(8, 6))
sns.heatmap(family_data.corr(), annot=True, cmap='PuOr', center=0)
plt.title('Relationship Between Family Features and Survival', pad=20, fontweight='bold')
plt.show()
```

```python
g = sns.FacetGrid(data, col='Sex', row='Survived', margin_titles=True)
g.map(sns.histplot, 'Age', bins=20, alpha=0.7, kde=True)
g.fig.suptitle('Age Distribution by Gender and Survival Status', y=1.05, fontweight='bold')
g.set_axis_labels('Age', 'Count')
plt.tight_layout()
plt.show()
```
