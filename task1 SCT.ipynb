```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
```

```python
zip_file = "C:\\Users\\amish\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_38144.zip"
csv_file = "API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"

with zipfile.ZipFile(zip_file) as z:
    with z.open(csv_file) as f:
        df = pd.read_csv(f, skiprows=4)
```

```python
india = df[df["Country Name"] == "India"]
years = [str(y) for y in range(2000, 2025)]
population = india[years].values.flatten()
```

```python

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.bar(years, population / 1e9, color='skyblue')
plt.title("India's Population (2000–2024)")
plt.xlabel("Year")
plt.ylabel("Population (in Billions)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()




population_in_billions = population / 1e9

plt.figure(figsize=(12, 6))
plt.hist(population_in_billions, bins=10, color='purple', edgecolor='black')
plt.title("Distribution of India's Population (2000–2024)")
plt.xlabel("Population (in Billions)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
